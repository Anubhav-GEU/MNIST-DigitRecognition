```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import cv2
from PIL import Image
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
from keras.datasets import mnist
from tensorflow.math import confusion_matrix
```

# Loading the MNIST data from keras dataset

```python
(X_train, Y_train), (X_test, Y_test) = mnist.load_data()
```

```python
type(X_train) # the data set is already converted in 1D Array
```

```python
# shape of the numpy array
print(X_train.shape, Y_train.shape, X_test.shape, Y_test.shape)
```

Training data = 60,000 images
Test data = 10,000 Images

Image dimension --> 28x28
GrayScale Image --> 1 channel

# printing the image from data

```python
print(X_train[9])
print("The shape of image is ",X_train[9].shape)
```

```python
# converting numpy array to image
plt.imshow(X_train[25])
plt.show()

# printing the corresponding label
print(Y_train[25])
```

```python
# unique values in Y_train

print(np.unique(Y_train))

# unique values in Y_test
print(np.unique(Y_test))
```

```python
# scaling the data
X_train = X_train/255 # 255 is the maximum value that is present in the data
X_test = X_test/255
```

# Printing the image from data after NORMALIZATION

```python
# converting numpy array to image
plt.imshow(X_train[25])
plt.show()

# printing the corresponding label
print(Y_train[25])
```

# Building the Neural Network - Simple Neural Network

```python
#setting up the layers of the Neural Network

model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(50, activation='relu'),
    keras.layers.Dense(50, activation='relu'),
    keras.layers.Dense(50, activation='relu'),
    keras.layers.Dense(50, activation='relu'),
    keras.layers.Dense(10, activation='sigmoid')
])
```

# Compiling the neural network

```python
model.compile(optimizer = 'adam',
             loss = 'sparse_categorical_crossentropy',
             metrics = ['accuracy']) 
```

# Training the neural network

```python
model.fit(X_train, Y_train, epochs=10)
```

# Model Evaluation : Accuracy on test data

```python
loss,accuracy = model.evaluate(X_test, Y_test)
print(accuracy*100,"%")
```

```python
print("Accuracy of test data:",accuracy*100,"%")
```

# Checking individual prediction made by the model

```python
y_pred = model.predict(X_test)
```

```python
print(y_pred.shape)
```

```python
print(y_pred[0])
```

model.predict() gives the prediction probability of each class for that datapoint 
here class is from 0 to 9

# Converting prediction probabilities to class label

```python
label_for_first_test_image = np.argmax(y_pred[0])
print(label_for_first_test_image)
```

```python
y_pred_labels = [np.argmax(i) for i in y_pred]
print(y_pred_labels)
```

y_test --> true labels 



y_pred_labels --> predicted labels

Confusion Matrix for predicted labels

```python
conf_mat = confusion_matrix(Y_test, y_pred_labels)
```

```python
print(conf_mat)
```

# Building heap map using conf_mat using seaborn as sns

```python
plt.figure(figsize=(10,7))
sns.heatmap(conf_mat, annot = True , fmt = 'd' , cmap = 'Blues')
plt.ylabel('True Labels')
plt.xlabel("Predicted Labels")
```
